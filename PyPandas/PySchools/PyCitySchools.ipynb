```python
#### PyCitySchools ###

# import dependencies here
import pandas as pd
import numpy as np
```

```python
# read school records
school_records = pd.read_csv("raw_data/schools_complete.csv")
school_records.head()
```

```python
# read student records
student_records = pd.read_csv("raw_data/students_complete.csv")
student_records.head()
```

```python
# Total Schools
total_schools = school_records['School ID'].count()

# Total Students
total_students = school_records['size'].sum()

# Total Budget
total_budget = school_records['budget'].sum()
```

```python
# Average Math Score
average_math_score = student_records['math_score'].mean()

# Average Reading Score
average_reading_score = student_records['reading_score'].mean()

# Number / percentage of students with passing match score
passing_math_count = student_records.loc[student_records['math_score'] >= 70]['name'].count()
passing_math_percentage = passing_math_count / total_students * 100

# Number / percentage of students with passing reading score
passing_reading_count = student_records.loc[student_records['reading_score'] >= 70]['name'].count()
passing_reading_percentage = passing_reading_count / total_students * 100

# Overall Passing Rate (Average of the above two)
overall_passing_rate = np.mean([passing_math_percentage, passing_reading_percentage])
```

```python
# High level snapshot of the district's key metrics
distrit_schools_summary = pd.DataFrame({'Total Schools': total_schools,
                                        'Total Students': total_students,
                                        'Total Budget': total_budget,
                                        'Average Math Score': average_math_score,
                                        'Average Reading Score': average_reading_score,
                                        '% Passing Math': passing_math_percentage,
                                        '% Passing Reading': passing_reading_percentage,
                                        'Overall Passing Rate': overall_passing_rate
                                       }, index=[0])

# reorder the columns
distrit_schools_summary = distrit_schools_summary[['Total Schools',
                                                   'Total Students',
                                                   'Total Budget', 
                                                   'Average Math Score',
                                                   'Average Reading Score',
                                                   '% Passing Math', 
                                                   '% Passing Reading',
                                                   'Overall Passing Rate'
                                                  ]]

print("District Summary:")
distrit_schools_summary
```
